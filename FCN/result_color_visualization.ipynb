```python
import os
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets 
import IPython.display as display
from ipywidgets import Box, Layout, Image
import cv2
import matplotlib.cm as cm
import scipy.misc as misc
%matplotlib inline
```

```python
gt1 = misc.imread('./example_result/gt_5.png')
inp1 = misc.imread('./example_result/inp_5.png')
pred1 = misc.imread('./example_result/pred_5.png')
gt2 = misc.imread('./example_result/gt_6.png')
inp2 = misc.imread('./example_result/inp_6.png')
pred2 = misc.imread('./example_result/pred_6.png')
```

```python
box_layout = Layout(display='flex',
                    flex_flow='row',
                    border='solid',
                   width='100%',)

wImg1 = widgets.Image(layout = widgets.Layout(border="solid"), width="30%") 
wImg2 = widgets.Image(layout = widgets.Layout(border="solid"), width="30%")

items1 = [wImg1, wImg2]
box1 = Box(children=items1)

tmpStream = cv2.imencode(".jpeg", inp1)[1].tostring()
wImg1.value = tmpStream

tmpStream = cv2.imencode(".jpeg", inp2)[1].tostring()
wImg2.value = tmpStream

display.display(box1)
```

```python
plt.figure()
plt.imshow(gt1, cmap=cm.Paired ,vmin=0, vmax=151)
plt.colorbar()
plt.savefig('example_result/gt_5_color.png', bbox_inches='tight')
plt.figure()
plt.imshow(gt2, cmap=cm.Paired ,vmin=0, vmax=151)
plt.colorbar()
plt.savefig('example_result/gt_6_color.png', bbox_inches='tight')
```

```python
plt.figure()
plt.imshow(pred1,cmap=cm.Paired ,vmin=0, vmax=151)
plt.colorbar()
plt.savefig('example_result/pred_5_color.png', bbox_inches='tight')
plt.figure()
plt.imshow(pred2,cmap=cm.Paired ,vmin=0, vmax=151)
plt.colorbar()
plt.savefig('example_result/pred_6_color.png', bbox_inches='tight')
```
